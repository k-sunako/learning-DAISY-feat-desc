# Dense DAISY feature description
https://scikit-image.org/docs/dev/auto_examples/features_detection/plot_daisy.htmlfrom

```python
from skimage.feature import daisy
from skimage import data
import matplotlib.pyplot as plt


img = data.camera()
step=180
radius=255
descs, descs_img = daisy(img, step=step, radius=radius, rings=2, histograms=6,
                         orientations=12, visualize=True)

fig, ax = plt.subplots()
ax.axis('off')
ax.imshow(descs_img)
descs_num = descs.shape[0] * descs.shape[1]
ax.set_title(f'img: {img.shape}, step: {step}, radius: {radius}, ')
plt.show()
```
